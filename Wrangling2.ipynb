```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("MOCK_DATA.csv")

df["original_marks"] = df["marks"]


df["marks"] = df["marks"].fillna(df["marks"].mean())


Q1 = df["marks"].quantile(0.25)
Q3 = df["marks"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = (df["marks"] < lower_bound) | (df["marks"] > upper_bound)
df["marks"] = np.where(outliers, np.median(df["marks"]), df["marks"])


df["log_marks"] = np.log(df["marks"] + 1)  


plt.figure(figsize=(12, 10)) 


plt.subplot(3, 2, 1)
sns.boxplot(y=df["original_marks"])
plt.title("Boxplot Before Outlier Handling")


plt.subplot(3, 2, 2)
sns.boxplot(y=df["marks"])
plt.title("Boxplot After Outlier Handling")


plt.subplot(3, 2, 3)
sns.histplot(df["original_marks"], kde=True, color="orange")
plt.title("Distplot Before Outlier Handling")


plt.subplot(3, 2, 4)
sns.histplot(df["marks"], kde=True, color="green")
plt.title("Distplot After Outlier Handling")


plt.subplot(3, 2, 5)
sns.histplot(df["log_marks"], kde=True, color="blue")
plt.title("Distplot After Log Transformation")

plt.tight_layout()
plt.show()
```
